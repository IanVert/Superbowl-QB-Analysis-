## Analysis of the Season Stats of NFL QBs.

### A comparison between the two QBs of Superbowl LVI 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10,8]
```

```python
data_passing = pd.read_csv('2021NFL_passing_data.csv')
data_passing = data_passing[(data_passing['G'] > 10) | (data_passing['Yds'] > 1000)]
data_passing.head()
```

### The dataset has been reduced to include only QBs that played more than 10 games and threw for more than 1000 Yards.

### The season data for the two QBs, Matthew Stafford of the LA Rams and Joe Burrow of the Cinncinati Bengals, is isolated.


```python
StaffordData = data_passing[data_passing['Player'] == 'Matthew Stafford']
BurrowData = data_passing[data_passing['Player'] == 'Joe Burrow']
```

## *Focus Area: Completion Percentage*

```python
plt.scatter(x = data_passing['Cmp%'], y = data_passing['TD'] , color = 'grey', label = 'Other QBs')
plt.scatter(x = StaffordData['Cmp%'], y = StaffordData['TD'] , color = 'Blue', label = 'Stafford')
plt.scatter(x = BurrowData['Cmp%'], y = BurrowData['TD'] , color = 'Orange', label = 'Burrow')

plt.xlabel('Completion Percentage', fontsize = 20)
plt.ylabel('Touchdowns', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(loc = 'upper left')
plt.savefig('CMP%_TD.png')
plt.show()
```

### **In the graph above it is shown that Burrow has less Touchdowns but has a better Completion Percentage**

```python
plt.scatter(x = data_passing['Cmp%'], y = data_passing['Y/C'] , color = 'grey', label = 'Other QBs')
plt.scatter(x = StaffordData['Cmp%'], y = StaffordData['Y/C'] , color = 'Blue', label = 'Stafford')
plt.scatter(x = BurrowData['Cmp%'], y = BurrowData['Y/C'] , color = 'Orange', label = 'Burrow')

plt.xlabel('Completion Percentage', fontsize = 20)
plt.ylabel('Net Yards Gained per Pass Completion', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(loc = 'upper left')
plt.savefig('CMP%_NYGPC.png')
plt.show()
```

### **Compared to Stafford, Burrow is throwing more accurate passes for more net yards per completion**

```python
plt.scatter(x = data_passing['Cmp%'], y = data_passing['Yds'] , color = 'grey', label = 'Other QBs')
plt.scatter(x = StaffordData['Cmp%'], y = StaffordData['Yds'] , color = 'Blue', label = 'Stafford')
plt.scatter(x = BurrowData['Cmp%'], y = BurrowData['Yds'] , color = 'Orange', label = 'Burrow')

plt.xlabel('Completion Percentage', fontsize = 20)
plt.ylabel('Yards Gained by Passing', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(loc = 'upper left')
plt.savefig('CMP%_Yds.png')
plt.show()
```

### **Burrow may have a higher Completion Percentage, but Stafford had more total yards during the season**

## *Focus Area: Interceptions*

```python
plt.scatter(x = data_passing['Int'], y = data_passing['TD'] , color = 'grey', label = 'Other QBs')
plt.scatter(x = StaffordData['Int'], y = StaffordData['TD'] , color = 'Blue', label = 'Stafford')
plt.scatter(x = BurrowData['Int'], y = BurrowData['TD'] , color = 'Orange', label = 'Burrow')

plt.xlabel('Interceptions', fontsize = 20)
plt.ylabel('Touchdowns', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(loc = 'upper left')
plt.savefig('INT_TD.png')
plt.show()
```

### **During the 2021 season, Stafford threw both more Interception and Touchdowns than Burrow**

```python
plt.scatter(x = data_passing['Int'], y = data_passing['Yds.1'] , color = 'grey', label = 'Other QBs')
plt.scatter(x = StaffordData['Int'], y = StaffordData['Yds.1'] , color = 'Blue', label = 'Stafford')
plt.scatter(x = BurrowData['Int'], y = BurrowData['Yds.1'] , color = 'Orange', label = 'Burrow')

plt.xlabel('Interceptions', fontsize = 20)
plt.ylabel('Yards Lost due to sacks', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(loc = 'upper left')
plt.savefig('INT_YDSLDS.png')
plt.show()
```

### **During the season, Stafford was tied for most interceptions and Burrow was tied for most yards lost due to sacks**

```python
plt.scatter(x = data_passing['Int'], y = data_passing['Yds'] , color = 'grey', label = 'Other QBs')
plt.scatter(x = StaffordData['Int'], y = StaffordData['Yds'] , color = 'Blue', label = 'Stafford')
plt.scatter(x = BurrowData['Int'], y = BurrowData['Yds'] , color = 'Orange', label = 'Burrow')

plt.xlabel('Interceptions', fontsize = 20)
plt.ylabel('Yards Gained by Passing', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(loc = 'upper left')
plt.savefig('INT_Yds.png')
plt.show()
```

### **Stafford threw for more yards but also more Interceptions than Burrow**

## *Conclusion:*

### **It is clear that during the 2021 Season Stafford and Burrow had different strengths and weaknesses. Stafford threw more yard than Burrow, despite Burrow having a better pass completion percentage. This could mean that Burrows pass completions occur with shorter yardage throws than Stafford, further analysis is needed. From the second focus area it is clear that Burrow  had more of an issue in the 2021 season with getting sacked than interceptions,and Stafford tied with throwing the most interceptions in the season.
